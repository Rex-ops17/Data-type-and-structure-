```python
Data Types and Structures Questions

 * Data structures are ways to organize and store data in a computer so that it can be accessed and modified efficiently. They are important for efficient data management, algorithm design, and optimizing program performance.

 * Mutable: Can be changed after creation. Example: list = [1, 2, 3] (can add/remove elements). Immutable: Cannot be changed after creation. Example: string = "hello" (cannot change individual characters, only create new strings).

 * Lists: Ordered, mutable collections of items. Enclosed in []. Tuples: Ordered, immutable collections of items. Enclosed in ().

 * Dictionaries store data as key-value pairs. Each key is unique and maps to a specific value.

 * To store a collection of unique items, where order doesn't matter, and for efficient set operations like union, intersection, and difference.

 * A string is an immutable sequence of characters. It differs from a list because a list is a mutable sequence of any data type.

 * Tuples ensure data integrity by being immutable. Once created, their elements cannot be changed, preventing accidental modification.

 * A hash table is a data structure that implements an associative array abstract data type, which maps keys to values. In Python, dictionaries are implemented using hash tables.

 * Yes, lists can contain different data types (e.g., [1, "hello", 3.14, True]).

 * Strings are immutable because once a string is created, its content cannot be altered. Any operation that appears to modify a string actually creates a new string.

 * Dictionaries offer faster lookup times by key, are useful for representing structured data (like records), and allow for more descriptive access to data via keys.

 * A tuple would be preferable over a list for representing fixed collections of related data, such as a coordinate point (x, y) or RGB color (red, green, blue), where the values should not change.

 * Sets handle duplicate values by automatically storing only unique elements. If you try to add a duplicate, it's ignored.

 * The "in" keyword checks for membership. For lists, it checks if an item exists within the list. For dictionaries, it checks if a key exists within the dictionary.

 * No, you cannot modify the elements of a tuple. Tuples are immutable, meaning their contents cannot be changed after creation.

 * A nested dictionary is a dictionary where the values themselves are other dictionaries. Example: {'user1': {'name': 'Alice', 'age': 30}, 'user2': {'name': 'Bob', 'age': 25}}. Use case: storing complex, hierarchical data.

 * The time complexity of accessing elements in a dictionary is O(1) on average (constant time).

 * Lists are preferred when the order of elements is important, when duplicates are allowed, when the collection needs to be frequently modified (adding/removing elements), and when accessing elements by index.

 * Dictionaries are considered unordered because, in older Python versions, the order of insertion was not guaranteed. While modern Python (3.7+) preserves insertion order, they are fundamentally designed for key-based lookups rather than ordered iteration, and their primary benefit isn't sequential access. This doesn't affect data retrieval by key, which remains efficient.

 * List: Data is retrieved by numerical index (position). Accessing list[0], list[1], etc. Dictionary: Data is retrieved by unique keys. Accessing dictionary['key_name'].


Practical Questions

# 1. Write a code to create a string with your name and print it.
my_name = "YourName" # Replace YourName with an actual name
print(my_name)

# 2. Write a code to find the length of the string "Hello World".
s = "Hello World"
print(len(s))

# 3. Write a code to slice the first 3 characters from the string "Python Programming".
s = "Python Programming"
print(s[0:3])

# 4. Write a code to convert the string "Hello" to uppercase.
s = "Hello"
print(s.upper())

# 5. Write a code to replace the word "apple" with "orange" in the string "I like apple".
s = "I like apple"
print(s.replace("apple", "orange"))

# 6. Write a code to create a list with numbers 1 to 5 and print it.
my_list = [1, 2, 3, 4, 5]
print(my_list)

# 7. Write a code to append the number 10 to the list [1, 2, 3, 4].
my_list = [1, 2, 3, 4]
my_list.append(10)
print(my_list)

# 8. Write a code to remove the number 3 from the list [1, 2, 3, 4, 5].
my_list = [1, 2, 3, 4, 5]
my_list.remove(3)
print(my_list)

# 9. Write a code to access the second element in the list ['a', 'b', 'c', 'd'].
my_list = ['a', 'b', 'c', 'd']
print(my_list[1])

# 10. Write a code to reverse the list [10, 20, 30, 40, 50].
my_list = [10, 20, 30, 40, 50]
my_list.reverse()
print(my_list)

# 11. Write a code to create a tuple with the elements 100, 200, 300 and print it.
my_tuple = (100, 200, 300)
print(my_tuple)

# 12. Write a code to access the second-to-last element of the tuple ('red', 'green', 'blue', 'yellow').
my_tuple = ('red', 'green', 'blue', 'yellow')
print(my_tuple[-2])

# 13. Write a code to find the minimum number in the tuple (10, 20, 5, 15).
my_tuple = (10, 20, 5, 15)
print(min(my_tuple))

# 14. Write a code to find the index of the element 'cat' in the tuple ('dog', 'cat', 'rabbit').
my_tuple = ('dog', 'cat', 'rabbit')
print(my_tuple.index('cat'))

# 15. Write a code to create a tuple containing three different fruits and check if "kiwi" is in it.
my_tuple = ('apple', 'banana', 'orange')
print("kiwi" in my_tuple)

# 16. Write a code to create a set with the elements 'a', 'b', 'c' and print it.
my_set = {'a', 'b', 'c'}
print(my_set)

# 17. Write a code to clear all elements from the set {1, 2, 3, 4, 5}.
my_set = {1, 2, 3, 4, 5}
my_set.clear()
print(my_set)

# 18. Write a code to remove the element 4 from the set {1, 2, 3, 4}.
my_set = {1, 2, 3, 4}
my_set.remove(4)
print(my_set)

# 19. Write a code to find the union of two sets {1, 2, 3} and {3, 4, 5}.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2))

# 20. Write a code to find the intersection of two sets {1, 2, 3} and {2, 3, 4}.
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print(set1.intersection(set2))

# 21. Write a code to create a dictionary with the keys "name", "age", and "city", and print it.
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print(my_dict)

# 22. Write a code to add a new key-value pair "country": "USA" to the dictionary {'name': 'John', 'age': 25}.
my_dict = {'name': 'John', 'age': 25}
my_dict['country'] = 'USA'
print(my_dict)

# 23. Write a code to access the value associated with the key "name" in the dictionary {'name': 'Alice', 'age': 30}.
my_dict = {'name': 'Alice', 'age': 30}
print(my_dict['name'])

# 24. Write a code to remove the key "age" from the dictionary {'name': 'Bob', 'age': 22, 'city': 'New York'}.
my_dict = {'name': 'Bob', 'age': 22, 'city': 'New York'}
del my_dict['age']
print(my_dict)

# 25. Write a code to check if the key "city" exists in the dictionary {'name': 'Alice', 'city': 'Paris'}.
my_dict = {'name': 'Alice', 'city': 'Paris'}
print('city' in my_dict)

# 26. Write a code to create a list, a tuple, and a dictionary and print them all.
my_list = [1, 2, 3]
my_tuple = (4, 5, 6)
my_dict = {"a": 7, "b": 8}
print(my_list)
print(my_tuple)
print(my_dict)

# 27. Write a code to create a list of 5 random numbers between 1 and 100, sort it in ascending order, and print the result.
import random
random_numbers = [random.randint(1, 100) for _ in range(5)]
random_numbers.sort()
print(random_numbers)

# 28. Write a code to create a list with strings and print the element at the third index.
my_list = ["apple", "banana", "cherry", "date"]
print(my_list[2])

# 29. Write a code to combine two dictionaries into one and print the result.
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
combined_dict = {**dict1, **dict2}
print(combined_dict)

# 30. Write a code to convert a list of strings into a set.
my_list = ["apple", "banana", "apple", "orange"]
my_set = set(my_list)
print(my_set)
```
